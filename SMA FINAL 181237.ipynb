```python
def clean_text(text):
    text=text.replace('\n','')
    text=re.sub(r'https?:\/\/.*[\r\n]*', '', text)
    text=re.sub("@\s*\S+","",text)
    text_cleaned=re.sub("pic.\S+","",text)
    
    return text_cleaned
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner']) #Loads the english model and we dont want prser or an ner so we will disable it
```

```python
stopwords=spacy.lang.en.stop_words.STOP_WORDS
```

```python
def final_clean(text):
    #text=re.sub('#\s*\S+','',text)
    text=re.sub(' +', ' ', text)
    text=text.strip().lower()
    document = nlp(text)
    doc_cleaned=[token.lemma_ for token in document if (token.text not in stopwords) and token.pos_ not in ['PUNCT','SPACE',"PRON","NUM"]]
    doc_cleaned=" ".join(doc_cleaned)

    return doc_cleaned
```

```python
tweet['cleaned_tweet']=tweet['cleaned_tweet'].apply(lambda x:re.sub('#\s*\S+','',x))
```

```python
tweet['final_tweet']=tweet['cleaned_tweet'].apply(lambda x:final_clean(x))
```

# What words come mostly frequently in Tweet - unigram, bigram and trigrams 

#unigrams

```python
tweets=tweet['final_tweet'].tolist()
```

```python
tweets[:8]
```

```python
def getMostCommon(reviews_list,topn=20):
    reviews=" ".join(reviews_list)
    tokenised_reviews=reviews.split(" ")
    
    
    freq_counter=Counter(tokenised_reviews)
    return freq_counter.most_common(topn)
    
```

```python
def plotMostCommonWords(reviews_list,topn=50,title="Positive Review",color="blue",axis=None):
    top_words=getMostCommon(reviews_list,topn=topn)
    data=pd.DataFrame()
    data['words']=[val[0] for val in top_words]
    data['freq']=[val[1] for val in top_words]
    if axis!=None:
        sns.barplot(y='words',x='freq',data=data,color=color,ax=axis).set_title(title+" top "+str(topn))
    else:
        sns.barplot(y='words',x='freq',data=data,color=color).set_title(title+" top "+str(topn))
```

```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 8,6 ## Sets the heigth and width of image
plotMostCommonWords(tweets,20,"most common words")
```

#bigram

```python
def generateNGram(text,n):
    tokens=text.split(" ")
    ngrams = zip(*[tokens[i:] for i in range(n)])
    return ["_".join(ngram) for ngram in ngrams]
```

```python
tweets_bigrams=[" ".join(generateNGram(review,2))for review in tweets]
```

```python
tweets_bigrams[:5]
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6 ## Sets the heigth and width of image
plotMostCommonWords(tweets_bigrams,20,"most common words")
```

#trigrams

```python
tweets_trigrams=[" ".join(generateNGram(review,3))for review in tweets]
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6 ## Sets the heigth and width of image
plotMostCommonWords(tweets_trigrams,20,"most common words")
```

# Using VADER or TextBlob identify the sentiment of the tweet. 

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser=SentimentIntensityAnalyzer()
```

```python
def sentiment_scores(sentence):
    scores=analyser.polarity_scores(sentence)
    return scores['compound']
```

```python
tweet['sentiments']=tweet['cleaned_tweet'].apply(lambda x:sentiment_scores(x))
```

```python
tweet.sample(5)
```

```python
def sentiment_category(scores):
    if scores >= 0.05:
        return "positive"
    elif scores <= -0.05:
        return "negative"
    else:
        return "neutral"
```

```python
tweet['sentiment_score']=tweet['sentiments'].apply(lambda x:sentiment_category(x))
```

```python
tweet.sample(10)
```

```python
tweet['sentiment_score'].value_counts()
```

#Create a dictionary using corpora

```python
import plotly.graph_objs as go
import os
import gensim
from gensim import corpora, models, similarities
import logging
import tempfile
from nltk.corpus import stopwords
from string import punctuation
from collections import OrderedDict
import warnings
warnings.filterwarnings("ignore")
import pyLDAvis.gensim
```

```python
corpus=[]
a=[]
for i in range(len(tweet['final_tweet'])):
        a=tweet['final_tweet'][i]
        corpus.append(a)
```

```python
corpus[0:10]
```

```python
TEMP_FOLDER = tempfile.gettempdir()
```

```python
print('Folder "{}" will be used to save temporary dictionary and corpus.'.format(TEMP_FOLDER))
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
# removing common words and tokenizing
list1 = ['RT','rt']
stoplist = stopwords.words('english') + list(punctuation) + list1
```

```python
texts = [[word for word in str(document).lower().split() if word not in stoplist] for document in corpus]
```

```python

dictionary = corpora.Dictionary(texts)
dictionary.save(os.path.join(TEMP_FOLDER, 'elon.dict'))  # store the dictionary, for future reference

#print(dictionary)
#print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
corpora.MmCorpus.serialize(os.path.join(TEMP_FOLDER, 'elon.mm'), corpus)  # store to disk, for later use
```

```python
tfidf = models.TfidfModel(corpus) # step 1 -- initialize a model
```

```python
corpus_tfidf = tfidf[corpus]  # step 2 -- use the model to transform vectors
```

```python
total_topics = 5
```

```python
lda = models.LdaModel(corpus, id2word=dictionary, num_topics=total_topics)
corpus_lda = lda[corpus_tfidf] # create a double wrapper over the original corpus: bow->tfidf->fold-in-lsi
```

```python
lda.show_topics(total_topics,5)
```

```python
data_lda = {i: OrderedDict(lda.show_topic(i,25)) for i in range(total_topics)}
#data_lda
```

```python
df_lda = pd.DataFrame(data_lda)
df_lda = df_lda.fillna(0).T
print(df_lda.shape)
```
